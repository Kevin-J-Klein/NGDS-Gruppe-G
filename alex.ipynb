```python
max = int(np.max(time_diff))
t = np.linspace(0, max, (2*max))

f_cubic = interpolate.interp1d(time_diff, prestah0, kind='cubic')
f_linear = interpolate.interp1d(time_diff, prestah0, kind = 'linear')

pressure_inter_cubic = f_cubic(t)
pressure_inter_linear = f_linear(t)
```

 7. Mithilfe der Ableitung soll nach grossen Änderungen in den Daten gesucht werden. 
   - So könnt ihr z.B. sagen: "Am Tag X hatten wir die grösste Änderungsrate des Luftdrucks".

```python
# Ableitung des Luftdrucks berechnen
dp_dt = np.gradient(prestah0, time_diff)

# Index des Maximums in der Ableitung finden
max_idx = np.argmax(dp_dt)

# Datum und Luftdruckwert zum Zeitpunkt des Maximums finden
max_date = time[max_idx]
max_pressure = prestah0[max_idx]

# Ergebnis ausgeben
print(f"Am {max_date} hatten wir die größte Änderungsrate des Luftdrucks ({dp_dt[max_idx]:.2f} hPa/s) und einen Luftdruck von {max_pressure:.2f} hPa.")
```
